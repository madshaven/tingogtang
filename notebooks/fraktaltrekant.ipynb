```python
import numpy as np
import plotly.express as px
```

```python
class Position:
    def __init__(self, x=0, y=0, a=(0, 0), b=(1, 0), c=(0.5, 1)):
        self.x, self.y = x, y
        self.a, self.b, self.c = a, b, c

    def zeno(self, x, y):
        # self.x, self.y = 0.5*(self.x + x), 0.5*(self.y + y)
        # self.x, self.y = self.x + 2*(x - self.x)/3, self.y + 2*(y - self.y)/3
        # self.x, self.y = self.x + 1*(x - self.x)/3, self.y + 1*(y - self.y)/3
        self.x, self.y = self.x + (x - self.x)/2, self.y + .55*(y - self.y)

    def zeno_a(self): self.zeno(*self.a)
    def zeno_b(self): self.zeno(*self.b)
    def zeno_c(self): self.zeno(*self.c)

    def move(self):
        choice = np.random.randint(0, 3)
        moves = [self.zeno_a, self.zeno_b, self.zeno_c]
        _ = moves[choice]()
        return self.x, self.y
    
    def moveNshow(self, n):
        places = [pos.move() for i in range(int(n))]
        x, y = zip(*places)
        px.scatter(x=x, y=y).show()
```
